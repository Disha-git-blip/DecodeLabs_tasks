```python
import random

def run_chatbot():
    print("Rule-Based AI Online")
    print("Type 'exit' to disconnect.\n")

    responses = {
        'hello': [
            'Hi there! Welcome to the system.',
            'Hello! Great to see you.',
            'Greetings, user! How can I assist?'
        ],
        'status': [
            'All deterministic guardrails are functioning normally.',
            'System is operating at 100% capacity.',
            'Logic engine is online and stable.'
        ],
        'help': [
            'I am a rule-based bot. Try saying hello, status, or bye.',
            'Need a hand? You can ask me for my status, or just say hi.',
        ],
        'bye': [
            'Goodbye! Safe travels.',
            'Shutting down connection. See you next time!',
            'Bye!'
        ]
    }

    fallbacks = [
        "I do not understand. Please try a different command.",
        "That command is not in my database.",
        "Error: Intent not recognized."
    ]

    while True:
        raw_input = input("You: ")
        clean_input = raw_input.lower().strip()

        if clean_input == 'exit':
            print("Bot: Initiating shutdown sequence... Goodbye.")
            break

        if clean_input in responses:
            reply = random.choice(responses[clean_input])
        else:
            reply = random.choice(fallbacks)

        print(f"Bot: {reply}")

run_chatbot()
```
